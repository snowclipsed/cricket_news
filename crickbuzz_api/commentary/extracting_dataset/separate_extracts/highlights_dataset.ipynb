```python
import json
import os

folder_path_1 = "/Users/aishwaryaravichandran/Downloads/crickbuzz_api/commentary/highlights/1st"

filename_1 = "66169.json"

file_path_1 = os.path.join(folder_path_1, filename_1)

with open(file_path_1, "r") as file:
    json_data = json.load(file)
```

```python
import pandas as pd

data_1 = [{"over_number": obj["overNumber"] if "overNumber" in obj else None, "commentary_data": obj["commText"], "timestamp": obj["timestamp"]} for obj in json_data]

data_1 = sorted(data_1, key=lambda x: x["timestamp"])

df1 = pd.DataFrame(data_1)

print(df1)
```

```python
folder_path_2 = "/Users/aishwaryaravichandran/Downloads/crickbuzz_api/commentary/highlights/2nd"

filename_2 = "66169.json"

file_path_2 = os.path.join(folder_path_2, filename_2)

with open(file_path_2, "r") as file:
    json_data = json.load(file)
```

```python
data = [{"over_number": obj["overNumber"] if "overNumber" in obj else None, "commentary_data": obj["commText"], "timestamp": obj["timestamp"]} for obj in json_data]

data = sorted(data, key=lambda x: x["timestamp"])

df2 = pd.DataFrame(data)

print(df2)
```

```python
appended_df = pd.concat([df1, df2], ignore_index=True)

appended_df.head(100)
```

```python
pattern_1 = r'\b\d{1,2}\.\d{2}\s*(?:AM|PM)?\b'

appended_df['commentary_data'] = appended_df['commentary_data'].str.replace(pattern_1, '', regex=True)

pattern_2 = r'\bB\S*\$'

appended_df['commentary_data'] = appended_df['commentary_data'].str.replace(pattern_2, '', regex=True)

pattern_3 = r'@'

appended_df['commentary_data'] = appended_df['commentary_data'].str.replace(pattern_3, '', regex=True)

appended_df.head(100)
```

```python
appended_df.to_csv('/Users/aishwaryaravichandran/Downloads/crickbuzz_api/commentary/dataset/66169_highlights.csv', index=False)
```
